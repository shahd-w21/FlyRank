## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
import os
print(os.path.exists("data/raw/content_refresh_anonymized.csv"))
```

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Evidence check: trend_direction is the categorical field the label is built from
print(df["trend_direction"].value_counts())

task_type = "scoring (via classification probability, used for ranking)"
print("Chosen task type:", task_type)
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
df["is_declining_label"] = (df["trend_direction"] == "down").astype(int)

# Check: what fraction of pages are labeled "declining"?
print(df["is_declining_label"].value_counts(normalize=True))
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
import numpy as np

def precision_at_k(y_true, scores, k=50):
    order = np.argsort(scores)[::-1][:k]
    return np.asarray(y_true)[order].mean()

# Demonstration check using a random score as a placeholder
demo_scores = np.random.rand(len(df))
print(f"Precision@50 (random baseline check): "
      f"{precision_at_k(df['is_declining_label'], demo_scores, 50):.3f}")
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
df = df[(df["impressions_90d"] > 0) & (df["content_age_days"] >= 90)]
df = df.drop_duplicates(subset="content_id")

print(f"Rows after filtering: {len(df)}")
df[["content_id", "client_id", "impressions_90d", "sessions_90d",
    "content_age_days", "trend_direction", "is_declining_label"]].head(10)
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
rule_flag = (df["impressions_90d"] < df["impressions_90d"].median()).astype(int)
agreement = (rule_flag == df["is_declining_label"]).mean()
print(f"Single-signal rule agreement with label: {agreement:.3f}")
```
